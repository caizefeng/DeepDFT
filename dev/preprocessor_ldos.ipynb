```python
import numpy as np
from numba import numba
import sys
```

```python
# script, dir_name, window_num = sys.argv
def submit_gen(dir_name):
    dir_name = 'Al_100_surface/' + dir_name
    window_num = 180

    pot_name = dir_name + '/POTPAVG.dat'
    submit_name = 'submit_ldos'
    submit_out_name = dir_name + '/submit'

    pot_array = np.loadtxt(pot_name, skiprows = 1, dtype = np.float64)
    vacuum_level = np.max(pot_array[:,1])

    energy_array = vacuum_level - 0.1 * np.arange(window_num-1, -1, -1)

    with open(submit_name,'r') as f:
        f_out = open(submit_out_name, 'w')
        for line_index, line in enumerate(f):
            if line_index == 23:
                energy_str_array = [str(i) for i in energy_array]
                loop_sentence = line.rstrip()
                for i in range(window_num):
                    loop_sentence += ' '
                    loop_sentence += energy_str_array[i]
                f_out.write(loop_sentence + '\n')
            else:
                f_out.write(line.rstrip() + '\n')
        f_out.close()
        
```

```python
import os

def parchg_incar_gen(dir_name):
#     dir_name = 'Al_100_surface/' + dir_name
    incar_name = dir_name + '/INCAR'  
    os.system("sed -i '$aISTART = 1' " + incar_name)
    os.system("sed -i '$aICHARG = 1' " + incar_name)
    os.system("sed -i '$aLPARD = .T.' " + incar_name)
    os.system("sed -i '$aLSEPB = .FALSE.' " + incar_name)
    os.system("sed -i '$aLSEPK = .FALSE.' " + incar_name)
    os.system("sed -i '$aEINT =' " + incar_name)
    
```

```python
def parchg_cal(dir_name_list):
all_dir_name = '' 
for dir_name in dir_name_list:
    all_dir_name += (' ' + dir_name) 
    parchg_incar_gen(dir_name)
    submit_gen(dir_name)
os.system('for i in' + all_dir_name + 
          '; do cd $i; sbatch submit ; cd $OLDPWD ; done')
```
